```python
import numpy as np
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("F:\\lab de IA\\kmeans\\xclara.csv")
f1 = df['V1'].values
f2 = df['V2'].values
k = 3
kmeans = KMeans(n_clusters=k).fit(df)
cent1 = kmeans.cluster_centers_[:,1]
cent2 = kmeans.cluster_centers_[:,0]
print(kmeans.cluster_centers_)
plt.scatter(f1,f2, c=kmeans.labels_, cmap='rainbow',marker='.')  
plt.scatter(cent2,cent1,color = "black",marker='x')
plt.grid(True)
plt.title("Nube de puntos")
```
